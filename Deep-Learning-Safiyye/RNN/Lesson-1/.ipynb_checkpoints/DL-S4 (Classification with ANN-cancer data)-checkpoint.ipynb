EarlyStop kullandigimizda model 1000 epoch' ta geldigi noktaya 95 epoch' ta geldi, overfit engellendi ve skorlar duzeldi.

learning_rate=0.005 yaptigimizda egitim suresi kisaldi ve yaklasik 45 epoch' ta egitim tamamlandi, skor düsmedi.

EarlyStop kullanmadan sadece dropout yaparak overfiti kesmeyi amacladik ve 1000 epoch calismasina ragmen model overfit olmadi. 

Simdi ise __BatchNormalization__ ile; aktivasyon fonksiyonundan önce ve sonra, EarlyStop' lu ve EarlyStop' suz modeller kurup degerlendirmeler yapacagiz. 

__!!!!__ BatchNormalization' in ReLu gibi aktivasyon fonksiyonlarindan önce, sigmoid ve tanh aktivasyon fonksiyonlarindan sonra kullanilmasi tavsiye edilir. __!!!__

BatchNormalization' i ReLu' dan önce ve sonra kullanarak sonuclarimizi kiyasladik.

EarlyStop uygulamadan ReLu' dan sonra BatchNormalization uyguladik. 200 epoch' tan sonra model exploding yapmaya basladi. Arastirmacilar, zaman icinde BatchNormalization isleminin __Gradient Exploding__' e sebep oldugunu görmüsler. Biz de bunu teyit etmis olduk.

EarlyStop uygulamadan ReLu' dan önce BatchNormalization uyguladik. Patlamalarin sıklığı azaldi fakat olan patlamalar daha büyük. Iki model ile BatchNormalization' in patlamalara sebep oldugunu teyit etmis olduk. 

EarlyStop uygulayarak ReLu' dan sonra BatchNormalization uyguladik. Sadece EarlyStop kullandigimizda 95 epoch seviyelerinde egitim tamamlanirken, modele BatchNormalization ekledigimizde 55 epoch' ta neredeyse ayni skorlari elde etmis olduk. Yani egitim hizimiz artti.

EarlyStop uygulayarak ReLu' dan önce BatchNormalization uyguladik. Egitim 95 epoch civari bitti, egitim cok hizlanmadi fakat skor daha yüksek çıktı.

__Gradient Clipping --->__ __Exploding Gradients__ problemleri ile mücadele etmek icin kullanilir. __clipnorm__ ve __clipvalue__ isminde iki adet hyperparametresi vardir. Bu hyperparametreler optimizasyon algoritmasinin icine tanimlanir. (Asagida ayrintili anlatim var)

__L2 Norm -->__ Distance ölçen norm fonksiyonudur. Direk olarak clipvalue degeri girilerek gradient' in alacagi min ve max degerler de belirtilebilir fakat biz burda clipnorm kullanmayi tercih ettik. BatchNormalization uyguladigimizda patlama olan modele __clipnorm=0.5__ degerini ekleyerek modeli tekrar calistirdik. Bu islem sonucunda patlamalar oldukca azaldi (Asagida görseller var)

## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('Cancer', axis=1)
y = df['Cancer'].values
```

```python
seed = 42
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.10, random_state = seed)
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Modelling & Model Performance

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose=1)
```

### Gradient Clipping

Gradient clipping can be used to alleviate the problem of exploding gradients. This is just scaling or rescaling our gradient vectors when they hit a threshold or maximum value.

Applying gradient clipping in TensorFlow models is quite straightforward. The only thing you need to do is pass the parameter to the optimizer function. All optimizers have a `clipnorm` and a `clipvalue` parameters that can be used to clip the gradients.

Incase of **clipnorm**, the l2 norm of the gradients is capped at the specified value.

While **clipvalue** caps the gradient values such that they don't exceed the specified value

**What is norm function?**

Informally, a norm is a function that accepts as input a vector from our vector space V and spits out a real number that tells us how big that vector is. In order for a function to qualify as a norm, it must first fulfill some properties, so that the results of this metrization process kind of “make sense”.

**What is L2 norm function?**

The L2 norm calculates the distance of the vector coordinate from the origin of the vector space. As such, it is also known as the Euclidean norm as it is calculated as the Euclidean distance from the origin. The result is a positive distance value.

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(30, input_dim = X_train.shape[1]))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dense(15))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dense(1, activation="sigmoid"))

opt = Adam(lr = 0.001, clipnorm=0.5)
model.compile(optimizer = opt, loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
#early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 15)
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          #callbacks = [early_stop]
         )
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

## Cross Validation

**Keras models** can be used in **scikit-learn** by wrapping them with the **KerasClassifier** or **KerasRegressor** class.

To use these wrappers you must define a function that creates and returns your Keras sequential model, then pass this function to the build_fn argument when constructing the KerasClassifier class.

The constructor for the KerasClassifier class can take default arguments that are passed on to the calls to model.fit(), such as the number of epochs and the batch size.

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_classifier():
    tf.random.set_seed(seed)
    classifier = Sequential()
    classifier.add(Dense(units = 30, activation = 'relu'))
    classifier.add(Dense(units = 15, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = "adam", loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
classifier_model = KerasClassifier(build_fn = build_classifier, batch_size = 32, epochs = 100, verbose=0)

scores = cross_validate(estimator = classifier_model, X = X_train, y = y_train,
                        scoring = ['accuracy', 'precision', 'recall', 'f1'], cv = 10)

df_scores = pd.DataFrame(scores, index = range(1, 11)).iloc[:, 2:]

df_scores
```

```python
df_scores_summary = pd.DataFrame({"score_mean" : df_scores.mean().values, "score_std" : df_scores.std().values},
                                 index = ["acc", "pre", "rec", "f1"])

df_scores_summary
```

## GridSearchCV

https://machinelearningmastery.com/grid-search-hyperparameters-deep-learning-models-python-keras/

```python
from tensorflow.keras.optimizers import Adam, Adadelta, RMSprop, Nadam
```

```python
def build_classifier(optimizer, learn_rate):
    tf.random.set_seed(seed)
    classifier = Sequential()
    classifier.add(Dense(units = 30, activation = 'relu'))
    classifier.add(Dense(units = 15, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer(learn_rate), loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
early_stop = EarlyStopping(monitor = "val_accuracy", mode = "auto", verbose = 1, patience = 15)
```

```python
classifier_model = KerasClassifier(build_fn = build_classifier, validation_split = 0.1, epochs = 200)

parameters = {'batch_size': [32, 64],
              'optimizer': [Adam, RMSprop, Adadelta, Nadam],
              'learn_rate': [0.001, 0.003, 0.005]}

grid_model = GridSearchCV(estimator = classifier_model,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs = -1,
                          verbose = 1)

grid_model.fit(X_train, y_train, callbacks = [early_stop])
```

```python
grid_model.best_score_
```
